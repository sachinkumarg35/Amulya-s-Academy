```python
#Printing stars in Hollow right angle triangle shape,
n = int(input("Enter the number: "))
for row in range(0, n):
    for col in range(0, n):
        if row==0 or col == n-1 or row == col:
            print("*", end="")
        else:
            print(end=" ")
    print()
```

```python
#Printing stars in Hollow Equilateral Triangle shape | Pyramid pattren.
n = int(input('Enter the number: '))
for row in range(1, n+1):
    for col in range(1, 2*n):
        if row == n or row+col == n+1 or col-row == n-1:
            print("*", end="")
        else:
            print(end = " ")
    print()
```
